```python
import spacy
import pandas as pd
import random
import time
import psutil
import os
from spacy.training.example import Example
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. CONFIGURAÇÃO DO EXPERIMENTO ---
MODEL_NAMES = ["Modelo Pequeno", "Modelo Médio", "Modelo Grande"]
TRAIN_DATA_PERCENTAGES = [0.1, 0.2, 0.5, 0.9, 1.0]

# --- 2. GERAÇÃO DE DADOS ---

def create_massive_training_dataset():
    """Gera um dataset de TREINO massivo usando templates."""
    # (O código desta função permanece o mesmo que você já tinha)
    data = []
    termos_pedido = ["pedido", "compra", "encomenda", "pacote", "item", "produto", "envio", "entrega"]
    acoes_rastreio = ["verificar", "rastrear", "checar", "consultar", "saber o status", "acompanhar", "ver o andamento"]
    for _ in range(500):
        frase = f"{random.choice(acoes_rastreio)} meu {random.choice(termos_pedido)}"
        data.append((frase, {"cats": {"verificar_status_pedido": 1.0, "falar_com_atendente": 0.0, "fora_de_escopo": 0.0}}))

    termos_atendente = ["humano", "atendente", "agente", "pessoa", "alguém", "especialista"]
    acoes_falar = ["falar com", "conversar com", "me transferir para", "chamar um"]
    for _ in range(500):
        frase = f"{random.choice(acoes_falar)} um {random.choice(termos_atendente)}"
        data.append((frase, {"cats": {"verificar_status_pedido": 0.0, "falar_com_atendente": 1.0, "fora_de_escopo": 0.0}}))

    perguntas_gerais = ["Qual a capital da Mongólia?", "Receita de bolo", "Notícias de hoje"]
    for _ in range(500):
        frase = random.choice(perguntas_gerais)
        data.append((frase, {"cats": {"verificar_status_pedido": 0.0, "falar_com_atendente": 0.0, "fora_de_escopo": 1.0}}))
    
    random.shuffle(data)
    return data

def create_manual_realistic_test_set():
    """
    Cria um conjunto de TESTE realista, escrito à mão, com ambiguidades
    e estruturas diferentes das do gerador de treino.
    """
    test_data = []
    
    # Intenção 1: Status do Pedido (com variações)
    status_phrases = [
        "kd meu pedido?", # Gíria
        "Não recebi o código de rastreio ainda.", # Indireto
        "Alguma novidade sobre a entrega do item 456?", # Mais natural
        "Queria só checar a compra que fiz ontem.",
        "Meu pacote está muito atrasado, o que aconteceu?"
    ]
    for frase in status_phrases:
        test_data.append((frase, {"cats": {"verificar_status_pedido": 1.0, "falar_com_atendente": 0.0, "fora_de_escopo": 0.0}}))

    # Intenção 2: Falar com Atendente (com frases mais complexas)
    atendente_phrases = [
        "Esse robô não resolve, me passa pra alguém de verdade.", # Mais complexo
        "Já tentei de tudo, preciso de uma pessoa pra me ajudar.", # Frase mais longa
        "Meu problema é sobre uma cobrança indevida, preciso de um especialista.", # Contexto específico
        "Chega de respostas automáticas, quero um humano.",
        "ajuda" # AMBÍGUO: Pode ser um pedido de ajuda geral ou para falar com alguém
    ]
    for frase in atendente_phrases:
        test_data.append((frase, {"cats": {"verificar_status_pedido": 0.0, "falar_com_atendente": 1.0, "fora_de_escopo": 0.0}}))

    # Intenção 3: Fora de Escopo (mas pode parecer "dentro do escopo")
    fora_escopo_phrases = [
        "Qual o status do meu relacionamento?", # Usa a palavra "status"
        "Quero falar sobre a vida.", # Usa a palavra "falar"
        "Vocês entregam pizza?", # Usa a palavra "entregam"
        "Obrigado pela ajuda.", # Neutro
        "Qual o telefone do suporte?", # Pode ser confundido com "falar com atendente"
    ]
    for frase in fora_escopo_phrases:
        test_data.append((frase, {"cats": {"verificar_status_pedido": 0.0, "falar_com_atendente": 0.0, "fora_de_escopo": 1.0}}))

    return test_data

# --- PREPARAÇÃO DOS DADOS ---
print("Gerando dataset de treino massivo...")
train_data = create_massive_training_dataset()

print("Criando dataset de teste realista e manual...")
test_data = create_manual_realistic_test_set()

print(f"Datasets criados: {len(train_data)} amostras de treino, {len(test_data)} amostras de teste (fixas).")


# --- FUNÇÃO DE TREINO E AVALIAÇÃO (permanece a mesma) ---
def train_and_evaluate(model_name: str, training_data: list, testing_data: list, total_train_size: int) -> dict:
    nlp = spacy.blank("pt")
    textcat = nlp.add_pipe("textcat")
    labels = list(training_data[0][1]['cats'].keys())
    for label in labels:
        textcat.add_label(label)

    process = psutil.Process(os.getpid())
    start_time = time.time()
    mem_before = process.memory_info().rss / (1024 * 1024)

    optimizer = nlp.begin_training()
    peak_mem = 0
    for i in range(10):
        random.shuffle(training_data)
        losses = {}
        for text, annotations in training_data:
            doc = nlp.make_doc(text)
            example = Example.from_dict(doc, annotations)
            nlp.update([example], sgd=optimizer, drop=0.1, losses=losses)
        current_mem = process.memory_info().rss / (1024 * 1024)
        if current_mem > peak_mem:
            peak_mem = current_mem

    training_time = time.time() - start_time
    training_peak_ram = peak_mem - mem_before
    
    texts_test, _ = zip(*testing_data)
    mem_inference = process.memory_info().rss / (1024 * 1024) - mem_before
    inference_start_time = time.time()
    preds = [nlp(text).cats for text in texts_test]
    inference_total_time = time.time() - inference_start_time
    avg_latency_ms = (inference_total_time / len(texts_test)) * 1000

    true_labels = [max(d['cats'], key=d['cats'].get) for _, d in testing_data]
    pred_labels = [max(p, key=p.get) for p in preds]
    accuracy = accuracy_score(true_labels, pred_labels)
    f1 = f1_score(true_labels, pred_labels, average='weighted')

    return {
        "model_name": model_name,
        "train_size_perc": len(training_data) / total_train_size * 100,
        "accuracy": accuracy,
        "f1_score": f1,
        "training_time_s": training_time,
        "training_peak_ram_mb": training_peak_ram,
        "inference_latency_ms": avg_latency_ms,
        "inference_ram_mb": mem_inference
    }

# --- EXECUÇÃO DO EXPERIMENTO ---
results = []
train_data_full_size = len(train_data)
print("\nIniciando o loop de experimentos com teste realista. Os resultados de acurácia serão menores.")

for model_ref_name in MODEL_NAMES:
    # Nota: Vamos usar porcentagens menores agora que o treino é maior e o teste é mais difícil
    for percentage in [0.01, 0.05, 0.1, 0.5, 1.0]: 
        subset_size = int(train_data_full_size * percentage)
        if subset_size < 1:
            continue
        # Criamos o subconjunto de treino a partir do dataset completo gerado
        train_subset = train_data[:subset_size]
        
        print(f"  Processando ref: {model_ref_name} com {len(train_subset)} amostras ({percentage*100:.0f}%)...")
        
        # O conjunto de teste (test_data) é sempre o mesmo conjunto manual
        result = train_and_evaluate(model_ref_name, train_subset, test_data, train_data_full_size)
        results.append(result)
        print(f"    -> Acurácia: {result['accuracy']:.4f} | F1-Score: {result['f1_score']:.4f} | Tempo: {result['training_time_s']:.2f}s")


# --- ANÁLISE E VISUALIZAÇÃO (permanece o mesmo) ---
df_results = pd.DataFrame(results)

print("\n--- Tabela Final de Resultados (Teste Realista) ---")
print(df_results)
df_results.to_csv("nlp_realistic_test_experiment_results.csv", index=False)
print("\nResultados salvos em 'nlp_realistic_test_experiment_results.csv'")

sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('Análise de Custo-Benefício com Teste Realista', fontsize=16)

sns.lineplot(data=df_results, x='train_size_perc', y='accuracy', hue='model_name', marker='o', ax=axes[0, 0])
axes[0, 0].set_title('Curva de Aprendizado (Performance)')
axes[0, 0].set_xlabel('% de Dados de Treino Utilizados')
axes[0, 0].set_ylabel('Acurácia no Conjunto de Teste')
axes[0, 0].set_ylim(0, 1.05)

# ... (o resto do código de plotagem é o mesmo) ...
sns.lineplot(data=df_results, x='inference_latency_ms', y='accuracy', hue='model_name', marker='o', ax=axes[0, 1])
axes[0, 1].set_title('Custo Operacional (Latência) vs. Performance')
axes[0, 1].set_xlabel('Latência de Inferência Média (ms)')
axes[0, 1].set_ylabel('Acurácia')

sns.lineplot(data=df_results, x='training_time_s', y='accuracy', hue='model_name', marker='o', ax=axes[1, 0])
axes[1, 0].set_title('Custo de Capital (Tempo de Treino) vs. Performance')
axes[1, 0].set_xlabel('Tempo de Treinamento (s)')
axes[1, 0].set_ylabel('Acurácia')

sns.lineplot(data=df_results, x='training_peak_ram_mb', y='accuracy', hue='model_name', marker='o', ax=axes[1, 1])
axes[1, 1].set_title('Custo de Capital (RAM de Treino) vs. Performance')
axes[1, 1].set_xlabel('Pico de Memória RAM no Treinamento (MB)')
axes[1, 1].set_ylabel('Acurácia')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("nlp_realistic_test_dashboard.png")
print("Painel de visualização salvo em 'nlp_realistic_test_dashboard.png'")
plt.show()
```
